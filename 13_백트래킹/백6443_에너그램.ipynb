```python
def perm(arr, r):
    results = []
    stack = [([], [False]*len(arr))]
    while stack:
        i_arr, visited = stack.pop()
        
        if(len(i_arr) == r):
            results.append("".join(i_arr))
            continue
        for i in range(len(arr)):
            if visited[i]: continue
            visited[i] = True
            stack.append((i_arr + [arr[i]], visited[:]))
            

    results = []
    def _perm(i_arr):
        if (len(i_arr) == r):
            results.append("".join(i_arr))
            return
        for i in range(len(arr)):
            if visited[i]: continue
            visited[i] = True
            _perm(i_arr + [arr[i]])
            visited[i] = False
    _perm([])
    return results
```

```python
from collections import defaultdict
n = int(input())

def perm(arr, r):
    visited = defaultdict(int)
    results = []
    def _perm(i_arr):
        if (len(i_arr) == r):
            results.append("".join(i_arr))
            return
        for i in range(len(arr)):
            if visited[i]: continue
            visited[i] = True
            _perm(i_arr + [arr[i]])
            visited[i] = False
    _perm([])
    return results

for i in range(n):
    word = list(sorted(input()))
    set_for_omiting_duplication = set()
    for anagram in perm(word, len(word)):
        if anagram in set_for_omiting_duplication:
            continue
        set_for_omiting_duplication.add(anagram)
        print(anagram)
```
