```python
import sys
#input = sys.stdin.readline
N = int(input())
N_APPLE = int(input())
RIGHT = "D"
LEFT = "L"
head_ds = (
    (0, 1),
    (1, 0),
    (0, -1),
    (-1, 0)
)
APPLES = set()
for _ in range(N_APPLE):
    temp = tuple(map(int, input().rstrip().split()))
    APPLES.add((temp[0] - 1, temp[1] - 1))
N_TURN_COMMANDS = int(input())
TURN_COMMANDS = []
for _ in range(N_TURN_COMMANDS):
    temp = tuple(input().rstrip().split())
    TURN_COMMANDS.append((int(temp[0]), temp[1]))

def move(head, tail, direction, maps):
    global APPLES
    next_r, next_c = head[0] + head_ds[direction][0], head[1] + head_ds[direction][1]
    if not(0<=next_r<N and 0<=next_c<N):
        return (True, (next_r, next_c), tail)
    if maps[next_r][next_c][0]:
        return (True, (next_r, next_c), tail)
    maps[head[0]][head[1]] = (True, direction)
    if (next_r, next_c) not in APPLES:
        new_tail = (tail[0] + head_ds[maps[tail[0]][tail[1]][1]][0], tail[1] + head_ds[maps[tail[0]][tail[1]][1]][1])
        maps[tail[0]][tail[1]] = (False, None)
        tail = new_tail
    else:
        APPLES.remove((next_r, next_c))
    return (False, (next_r, next_c), tail)
        
def get4n(now, d):
    result = now+d
    if result >=4:
        return 0
    elif result <0:
        return 3
    else:
        return result
i_command = 0
head = (0, 0)
tail = (0, 0)
direction = 0
maps = [[(False, None)]*N for _ in range(N)]
maps[0][0] = (True, 0)
for t in range(1, 10001):
    is_colide, head, tail = move(head, tail, direction, maps)
    #print(is_colide, head, tail)
    if is_colide:
        break
    if i_command < N_TURN_COMMANDS and TURN_COMMANDS[i_command][0] == t:
        command = TURN_COMMANDS[i_command][1]
        i_command+=1
        if command == 'D':
            direction = get4n(direction, 1)
        else:
            direction = get4n(direction, -1)
print(t)
```
