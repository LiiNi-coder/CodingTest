```python
"""
N, M = map(int, input().split())
maps = []
for n in range(N):
    temp = input()
    for m in range(M):
        if temp[m] == "B":
            blue = (n, m)
        if temp[m] == "R":
            red = (n, m)
        if temp[m] == "O":
            hole = (n, m)
    maps.append(list(temp))
"""
raw_input = """10 10
##########
#R#...##B#
#...#.##.#
#####.##.#
#......#.#
#.######.#
#.#....#.#
#.#.##...#
#O..#....#
##########
"""
raw_input = raw_input.splitlines()
N, M = map(int, raw_input[0].split())
maps = []
for n, line in enumerate(raw_input[1:]):
    temp = line
    for m in range(M):
        if temp[m] == "B":
            blue = (n, m)
        if temp[m] == "R":
            red = (n, m)
        if temp[m] == "O":
            hole = (n, m)
    maps.append(list(temp))

drdc = {0: (0, 1), 1:(1, 0), 2:(0, -1), 3:(-1, 0)}
def getNextDirections(b1, b2):
    possible_directions = []
    for i, temp in drdc.items():
        dr, dc = temp
        b1_next = b1[0] + dr, b1[1] + dc
        b2_next = b2[0] + dr, b2[1] + dc
        if maps[b1_next[0]][b1_next[1]] == "#" and maps[b2_next[0]][b2_next[1]] == "#":
            continue
        possible_directions.append(i)
    return possible_directions
def getNextPoints(b1, b2, direction):
    # b1, b2 모두 방향대로 보냄
    results_points = []
    results_flags = []
    for b in (b1, b2):
        flag = None # 0 : wall, 1: hole
        iter_b = list(b)
        is_moved = False
        while True:
            next_b = [iter_b[0] + drdc[direction][0], iter_b[1] + drdc[direction][1]]
            if maps[next_b[0]][next_b[1]] == "#":
                flag = 0
                break
            elif maps[next_b[0]][next_b[1]] == "O":
                iter_b = [next_b[0], next_b[1]]
                flag = 1
                break
            iter_b = [next_b[0], next_b[1]]
        results_points.append(tuple(iter_b))
        results_flags.append(flag)
        is_moved = True
    # b1과 b2의 좌표가 겹쳐있음 -> 그 이전 위치관계 파악해서 재조정
    if results_points[0] != results_points[1]:
        return (tuple(results_points[0]), tuple(results_points[1]), tuple(results_flags))
    if not is_moved:
        return (b1, b2, (0, 0))
    if tuple(results_flags) != (0, 0):
        return (b1, b2, tuple(results_flags))
    if direction == 0:#->
        if b1[1] < b2[1]:
            results_points[0] = (results_points[1][0], results_points[1][1] - 1, )
        else:
            results_points[1] = (results_points[0][0], results_points[0][1] - 1, )
    elif direction == 1:
        if b1[0] < b2[0]:
            results_points[0] = (results_points[1][0] - 1, results_points[1][1], )
        else:
            results_points[1] = (results_points[0][0] - 1, results_points[0][1], )
    elif direction == 2:
        if b1[1] < b2[1]:
            results_points[1] = (results_points[1][0], results_points[1][1] + 1, )
        else:
            results_points[0] = (results_points[0][0], results_points[0][1] + 1, )
    elif direction == 3:
        if b1[0] < b2[0]:
            results_points[1] = (results_points[1][0] + 1, results_points[1][1], )
        else:
            results_points[0] = (results_points[0][0] + 1, results_points[0][1], )
    return (tuple(results_points[0]), tuple(results_points[1]), tuple(results_flags))
    # 아니면 바로 끝
    
from collections import deque 
q = deque()
q.append((red, blue, 0))
answer = []
while q:
    #print(q)
    red, blue, count = q.popleft()
    #print(red, blue, count)
    #print("---")
    if count>=10:
        continue
    for now_d in getNextDirections(red, blue):
        n_red, n_blue, what_is_exit = getNextPoints(red, blue, now_d)
        if what_is_exit == (0, 0):
            q.append((n_red, n_blue, count + 1))
        elif what_is_exit == (1, 0):
            answer.append((count+1, n_red, n_blue))   
if len(answer) >= 1:
    print(min(map(lambda x:x[0], answer)))
else:
    print(-1)
```
