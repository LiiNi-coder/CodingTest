```python
raw = """6 7 10
.......
...O...
....O..
.......
OO.....
OO.....
"""
lines = raw.splitlines()
R, C, N = map(int, lines[0].split())
maps = [[0]*C for i in range(R)]
for r, line in enumerate(lines[1:]):
    for c in range(C):
        if line[c] == "O":
            maps[r][c] = 2
```
